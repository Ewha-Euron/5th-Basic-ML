```python
from nltk import word_tokenize, sent_tokenize

#ì—¬ëŸ¬ê°œì˜ ë¬¸ì¥ìœ¼ë¡œ ëœ ì…ë ¥ ë°ì´í„°ë¥¼ ë¬¸ì¥ë³„ë¡œ ë‹¨ì–´ í† í°í™” ë§Œë“œëŠ” í•¨ìˆ˜ ìƒì„±
def tokenize_text(text):

    #ë¬¸ì¥ë³„ë¡œ ë¶„ë¦¬ í† í°
    sentences = sent_tokenize(text)
    #ë¶„ë¦¬ëœ ë¬¸ì¥ë³„ ë‹¨ì–´ í† í°í™”
    word_tokens = [word_tokenize(sentence) for sentence in sentences]
    return word_tokens

#ì—¬ëŸ¬ ë¬¸ì¥ë“¤ì— ëŒ€í•´ ë¬¸ì¥ë³„ ë‹¨ì–´ í† í°í™” ìˆ˜í–‰
word_tokens = tokenize_text(text_sample)
print(type(word_tokens),len(word_tokens))
print(word_tokens)
```

```python
import nltk

stopwords = nltk.corpus.stopwords.words('english')
all_tokens = []
#ìœ„ ì˜ˆì œì—ì„œ 3ê°œì˜ ë¬¸ì¥ë³„ë¡œ ì–»ì€ word_tokens list ì— ëŒ€í•´ ìŠ¤í†± ì›Œë“œë¥¼ ì œê±°í•˜ëŠ” ë°˜ë³µë¬¸
for sentence in word_tokens:
    filtered_words=[]
    #ê°œë³„ ë¬¸ì¥ë³„ë¡œ í† í°í™”ëœ ë¬¸ì¥ listì— ëŒ€í•´ ìŠ¤í†± ì›Œë“œë¥¼ ì œê±°í•˜ëŠ” ë°˜ë³µë¬¸
    for word in sentence:
        #ì†Œë¬¸ìë¡œ ëª¨ë‘ ë³€í™˜
        word = word.lower()
        #í† í°í™”ëœ ê°œë³„ ë‹¨ì–´ê°€ ìŠ¤í†± ì›Œë“œì˜ ë‹¨ì–´ì— í¬í•¨ë˜ì§€ ì•Šìœ¼ë©´ word_tokensì— ì¶”ê°€
        if word not in stopwords:
            filtered_words.append(word)
    all_tokens.append(filtered_words)

print(all_tokens)
```

```python
import re

#<br> html íƒœê·¸ëŠ” replace í•¨ìˆ˜ë¡œ ê³µë°±ìœ¼ë¡œ ë³€í™˜
review_df['review'] = review_df['review'].str.replace('<br />',' ')

#íŒŒì´ì¬ì˜ ì •ê·œ í‘œí˜„ì‹ ëª¨ë“ˆì¸ reë¥¼ ì´ìš©í•˜ì—¬ ì˜ì–´ ë¬¸ìì—´ì´ ì•„ë‹Œ ë¬¸ìëŠ” ëª¨ë‘ ê³µë°±ìœ¼ë¡œ ë³€í™˜
review_df['review'] = review_df['review'].apply( lambda x : re.sub("[^a-zA-Z]", " ", x) )
```

```python
from sklearn.model_selection import train_test_split

class_df = review_df['sentiment']
feature_df = review_df.drop(['id','sentiment'], axis=1, inplace=False)

X_train, X_test, y_train, y_test= train_test_split(feature_df, class_df, test_size=0.3, random_state=156)

X_train.shape, X_test.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score

#ìŠ¤í†± ì›Œë“œëŠ” English, filtering, ngramì€ (1,2)ë¡œ ì„¤ì •í•´ CountVectorizationìˆ˜í–‰.
#LogisticRegressionì˜ CëŠ” 10ìœ¼ë¡œ ì„¤ì •.
pipeline = Pipeline([
    ('cnt_vect', CountVectorizer(stop_words='english', ngram_range=(1,2) )),
    ('lr_clf', LogisticRegression(solver='liblinear', C=10))])

#Pipeline ê°ì²´ë¥¼ ì´ìš©í•˜ì—¬ fit(), predict()ë¡œ í•™ìŠµ/ì˜ˆì¸¡ ìˆ˜í–‰. predict_proba()ëŠ” roc_auc ë•Œë¬¸ì— ìˆ˜í–‰.
pipeline.fit(X_train['review'], y_train)
pred = pipeline.predict(X_test['review'])
pred_probs = pipeline.predict_proba(X_test['review'])[:,1]

print('ì˜ˆì¸¡ ì •í™•ë„ëŠ” {0:.4f}, ROC-AUCëŠ” {1:.4f}'.format(accuracy_score(y_test ,pred), roc_auc_score(y_test, pred_probs)))
```

```python
#ìŠ¤í†± ì›Œë“œëŠ” english, filtering, ngramì€ (1,2)ë¡œ ì„¤ì •í•´ TF-IDF ë²¡í„°í™” ìˆ˜í–‰.
#LogisticRegressionì˜ CëŠ” 10ìœ¼ë¡œ ì„¤ì •.
pipeline = Pipeline([
    ('tfidf_vect', TfidfVectorizer(stop_words='english', ngram_range=(1,2) )),
    ('lr_clf', LogisticRegression(solver='liblinear', C=10))])

pipeline.fit(X_train['review'], y_train)
pred = pipeline.predict(X_test['review'])
pred_probs = pipeline.predict_proba(X_test['review'])[:,1]

print('ì˜ˆì¸¡ ì •í™•ë„ëŠ” {0:.4f}, ROC-AUCëŠ” {1:.4f}'.format(accuracy_score(y_test ,pred), roc_auc_score(y_test, pred_probs)))
```

sentiwordnetì„ ì´ìš©í•œ ì˜í™” ê°ìƒí‰ ê°ìƒ ë¶„ì„

```python
from nltk.corpus import wordnet as wn

#ê°„ë‹¨í•œ NTLK PennTreebank Tagë¥¼ ê¸°ë°˜ìœ¼ë¡œ WordNetê¸°ë°˜ì˜ í’ˆì‚¬ Tagë¡œ ë³€í™˜
def penn_to_wn(tag):
    if tag.startswith('J'):
        return wn.ADJ
    elif tag.startswith('N'):
        return wn.NOUN
    elif tag.startswith('R'):
        return wn.ADV
    elif tag.startswith('V'):
        return wn.VERB
    return
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import sentiwordnet as swn
from nltk import sent_tokenize, word_tokenize, pos_tag

def swn_polarity(text):
    #ê°ì„± ì§€ìˆ˜ ì´ˆê¸°í™”
    sentiment = 0.0
    tokens_count = 0

    lemmatizer = WordNetLemmatizer()
    raw_sentences = sent_tokenize(text)
    #ë¶„í•´ëœ ë¬¸ì¥ë³„ë¡œ ë‹¨ì–´ í† í° -> í’ˆì‚¬ íƒœê¹… í›„ì— SentiSynset ìƒì„± -> ê°ì„± ì§€ìˆ˜ í•©ì‚°
    for raw_sentence in raw_sentences:
        #NTLK ê¸°ë°˜ì˜ í’ˆì‚¬ íƒœê¹… ë¬¸ì¥ ì¶”ì¶œ
        tagged_sentence = pos_tag(word_tokenize(raw_sentence))
        for word , tag in tagged_sentence:

            #WordNet ê¸°ë°˜ í’ˆì‚¬ íƒœê¹…ê³¼ ì–´ê·¼ ì¶”ì¶œ
            wn_tag = penn_to_wn(tag)
            if wn_tag not in (wn.NOUN , wn.ADJ, wn.ADV):
                continue
            lemma = lemmatizer.lemmatize(word, pos=wn_tag)
            if not lemma:
                continue
            #ì–´ê·¼ì„ ì¶”ì¶œí•œ ë‹¨ì–´ì™€ WordNet ê¸°ë°˜ í’ˆì‚¬ íƒœê¹…ì„ ì…ë ¥í•´ Synset ê°ì²´ë¥¼ ìƒì„±.
            synsets = wn.synsets(lemma , pos=wn_tag)
            if not synsets:
                continue
            #sentiwordnetì˜ ê°ì„± ë‹¨ì–´ ë¶„ì„ìœ¼ë¡œ ê°ì„± synset ì¶”ì¶œ
            #ëª¨ë“  ë‹¨ì–´ì— ëŒ€í•´ ê¸ì • ê°ì„± ì§€ìˆ˜ëŠ” +ë¡œ ë¶€ì • ê°ì„± ì§€ìˆ˜ëŠ” -ë¡œ í•©ì‚°í•´ ê°ì„± ì§€ìˆ˜ ê³„ì‚°.
            synset = synsets[0]
            swn_synset = swn.senti_synset(synset.name())
            sentiment += (swn_synset.pos_score() - swn_synset.neg_score())
            tokens_count += 1

    if not tokens_count:
        return 0

    #ì´ scoreê°€ 0 ì´ìƒì¼ ê²½ìš° ê¸ì •(Positive) 1, ê·¸ë ‡ì§€ ì•Šì„ ê²½ìš° ë¶€ì •(Negative) 0 ë°˜í™˜
    if sentiment >= 0 :
        return 1

    return 0
```

```python
review_df['preds'] = review_df['review'].apply( lambda x : swn_polarity(x) )
y_target = review_df['sentiment'].values
preds = review_df['preds'].values
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
from sklearn.metrics import recall_score, f1_score, roc_auc_score
import numpy as np

print(confusion_matrix( y_target, preds))
print("ì •í™•ë„:", np.round(accuracy_score(y_target , preds), 4))
print("ì •ë°€ë„:", np.round(precision_score(y_target , preds),4))
print("ì¬í˜„ìœ¨:", np.round(recall_score(y_target, preds), 4))
```

VADERë¥¼ ì´ìš©í•œ ê°ìƒë¶„ì„

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

senti_analyzer = SentimentIntensityAnalyzer()
senti_scores = senti_analyzer.polarity_scores(review_df['review'][0])
print(senti_scores)
```

```python
def vader_polarity(review,threshold=0.1):
    analyzer = SentimentIntensityAnalyzer()
    scores = analyzer.polarity_scores(review)

    #compound ê°’ì— ê¸°ë°˜í•˜ì—¬ threshold ì…ë ¥ê°’ë³´ë‹¤ í¬ë©´ 1, ê·¸ë ‡ì§€ ì•Šìœ¼ë©´ 0ì„ ë°˜í™˜
    agg_score = scores['compound']
    final_sentiment = 1 if agg_score >= threshold else 0
    return final_sentiment

#apply lambdaì‹ì„ ì´ìš©, ë ˆì½”ë“œë³„ë¡œ vader_polarity()ë¥¼ ìˆ˜í–‰í•˜ê³  ê²°ê³¼ë¥¼ 'vader_preds'ì— ì €ì¥
review_df['vader_preds'] = review_df['review'].apply( lambda x : vader_polarity(x, 0.1) )
y_target = review_df['sentiment'].values
vader_preds = review_df['vader_preds'].values

print(confusion_matrix( y_target, vader_preds))
print("ì •í™•ë„:", np.round(accuracy_score(y_target , vader_preds),4))
print("ì •ë°€ë„:", np.round(precision_score(y_target , vader_preds),4))
print("ì¬í˜„ìœ¨:", np.round(recall_score(y_target, vader_preds),4))
```
