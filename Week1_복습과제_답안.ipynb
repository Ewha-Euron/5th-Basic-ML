### ì—°ìŠµë¬¸ì œ 2

def calc_many(choice, *args) í•¨ìˆ˜ì—ì„œ
- choiceê°€ `sum` ì´ë©´ **í•©**ì„ êµ¬í•´ì„œ return í•©ë‹ˆë‹¤.
- choice ê°€ `mul` ì´ë©´ **ê³±**í•œ ê²°ê³¼ë¥¼ return í•©ë‹ˆë‹¤.

í•¨ìˆ˜ë¥¼ êµ¬í˜„í•˜ì„¸ìš”

````python
# ì½”ë“œë¥¼ ì…ë ¥í•´ ì£¼ì„¸ìš”
def calc_many(choice, *args):
    # ì½”ë“œì…ë ¥
    ret = 0
    if choice == 'mul':
        ret = 1
    for i in args:
        if choice == 'sum':
            ret += i
        elif choice == 'mul':
            ret *= i
    return ret
````

````python
# ê²°ê³¼: 10
print(calc_many('sum', 0, 1, 2, 3, 4))
# ê²°ê³¼: 120
print(calc_many('mul', 1, 2, 3, 4, 5))
````
